# **Introduction au projet et aux objectifs**

Ce projet consiste à appliquer différentes techniques de prétraitement, de réduction de dimensionnalité, de classification et de clustering en machine learning sur un dataset que nous avons chargé.

Le dataset 'bienetre.csv' contient des variables explicatives ainsi qu'une variable cible qui représente les classes à prédire. Nous allons explorer les données, appliquer des techniques de normalisation et de réduction de dimensionnalité, puis tester plusieurs modèles de classification supervisée et non supervisée.

Les étapes principales de ce projet sont les suivantes :
1. Exploration du dataset et analyse descriptive.
2. Normalisation des données pour garantir une échelle uniforme des caractéristiques.
3. Réduction de la dimensionnalité avec PCA pour faciliter la visualisation.
4. Application de modèles de classification supervisée (KNN, régression logistique, arbre de décision).
5. Application de KMeans pour le clustering non supervisé.
6. Optimisation des hyperparamètres des modèles.
7. Validation croisée pour évaluer la stabilité et la performance des modèles.

L'objectif est d'évaluer la robustesse des modèles, d'optimiser leurs paramètres et de discuter des résultats obtenus.

# 1. Importation des Bibliothèques et Chargement et exploration du Dataset



```python
# Importation des librairies essentielles
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D

# Pour le prétraitement et la réduction de dimension
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Pour la modélisation et évaluation
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report, accuracy_score, auc, roc_curve, silhouette_score

# Pour le clustering
from sklearn.cluster import KMeans
```

```python
# Configuration de base pour les graphiques
sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (10, 6) # Taille par défaut des graphiques

# Chargement du dataset (bienetre.csv)
data = pd.read_csv('bienetre.csv')
```

```python
# Vue d'ensemble et description du dataset
print("Dimensions du dataset :", data.shape)
display(data.head())

print("\nRésumé statistique :")
display(data.describe())

print("\nValeurs manquantes par colonne :")
print(data.isnull().sum())

```

```python
# Matrice de Corrélation après Standardisation

# Séparation des variables explicatives et de la cible
target_column = 'target'  # Nom de la colonne cible
X = data.drop(columns=[target_column])  # Variables explicatives
y = data[target_column]  # Variable cible

# Standardisation des données
scaler = StandardScaler()
X_standardized = scaler.fit_transform(X)

# Création d'un DataFrame avec les données standardisées
X_standardized_df = pd.DataFrame(X_standardized, columns=X.columns)

# Calcul de la matrice de corrélation sur les données standardisées
correlation_matrix = X_standardized_df.corr()

# Affichage de la matrice de corrélation
plt.figure(figsize=(12, 12))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Matrice de Corrélation (Données Standardisées)")
plt.show()
```

# 2. Normalisation des Données

```python
# Comparaison visuelle de la distribution de plusieurs variables avant et après normalisation
# On utilise 'taille' et 'age' pour observer l'effet du centrage et de la réduction.

variables = ['taille', 'age']  # Liste des variables à comparer

for colonne in variables:
    # Affichage de la distribution avant normalisation
    plt.figure(figsize=(10, 6))
    sns.histplot(data[colonne], kde=True, color='skyblue')
    plt.title(f"Distribution de '{colonne}' avant normalisation")
    plt.xlabel(colonne)
    plt.ylabel("Fréquence")
    plt.show()

    # Normalisation des données (exclusion de la cible 'target')
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(data.drop(columns=['target']))

    # Récupération de l'indice de la colonne actuelle parmi les features
    col_index = data.drop(columns=['target']).columns.get_loc(colonne)

    # Affichage de la distribution après normalisation
    plt.figure(figsize=(10, 6))
    sns.histplot(X_scaled[:, col_index], kde=True, color='salmon')
    plt.title(f"Distribution de '{colonne}' après normalisation")
    plt.xlabel(colonne)
    plt.ylabel("Fréquence")
    plt.show()
```

# 3. Réduction de Dimensionnalité avec la PCA

```python
# Visualisation 2D avec PCA en fixant 2 composantes principales
# Pour faciliter la visualisation, on projette les données sur un plan en 2 dimensions.
pca_2 = PCA(n_components=2)

# Normalisation des données en exluant la variable cible ' target '
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# On suppose que la colonne 'target' est présente et qu'elle ne doit pas être normalisée
X_scaled = scaler.fit_transform(data.drop(columns=['target']))
X_pca_2 = pca_2.fit_transform(X_scaled)
```

```python
# Affichage de la variance expliquée par composante et de la variance cumulée
print("Variance expliquée par composante (2 composantes) :", pca_2.explained_variance_ratio_)
print("Variance cumulée (2 composantes) :", pca_2.explained_variance_ratio_.cumsum())
```

```python
# Visualisation 2D avec coloration par classe cible
y = data['target']
plt.figure()
sns.scatterplot(x=X_pca_2[:, 0], y=X_pca_2[:, 1], hue=y, palette='Set2')
plt.title("Projection PCA (2 composantes)")
plt.xlabel("Composante 1")
plt.ylabel("Composante 2")
plt.show()
```

```python
# Visualisation 3D avec PCA (n_components=3)
# Je projettes les données dans un espace 3D pour observer leur répartition selon trois axes principaux.

# Appliquer la PCA pour obtenir 3 composantes
pca_3 = PCA(n_components=3)
X_pca_3 = pca_3.fit_transform(X_scaled)

# Affichage de la variance expliquée par chaque composante et de la variance cumulée pour 3 composantes
print("Variance expliquée par composante (3 composantes):", pca_3.explained_variance_ratio_)
print("Variance cumulée (3 composantes):", pca_3.explained_variance_ratio_.cumsum())

# Création d'une visualisation 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(X_pca_3[:, 0], X_pca_3[:, 1], X_pca_3[:, 2], c=y, cmap='Set2')
ax.set_title("Projection PCA 3D (3 composantes)")
ax.set_xlabel("Composante 1")
ax.set_ylabel("Composante 2")
ax.set_zlabel("Composante 3")
fig.colorbar(scatter, ax=ax, label='Classe')
plt.show()
```

```python
# Normalisation des données (en excluant la variable cible 'target')
scaler = StandardScaler()
X_scaled = scaler.fit_transform(data.drop(columns=['target']))

# Détermination du nombre optimal de composantes (conserver 95% de la variance)
# On applique PCA avec n_components=0.95 pour savoir combien de composantes sont nécessaires
pca_95 = PCA(n_components=0.95)
X_pca_95 = pca_95.fit_transform(X_scaled)

# Affichage de la variance cumulée pour visualiser le seuil de 95%
plt.figure(figsize=(10, 6))
plt.plot(np.cumsum(pca_95.explained_variance_ratio_), marker='o')
plt.xlabel('Nombre de composantes')
plt.ylabel('Variance cumulée')
plt.title('Courbe de variance cumulée (PCA, 95% de variance)')
plt.axhline(y=0.95, color='r', linestyle='--', label='Seuil 95%')
plt.legend()
plt.show()

# Séparation en jeu d'entraînement et de test avec les données transformées par PCA
X_train_pca, X_test_pca, y_train, y_test = train_test_split(
    X_pca_95, data['target'], test_size=0.2, random_state=42, stratify=data['target']
)

print("Dimensions du jeu d'entraînement :", X_train_pca.shape)
print("Dimensions du jeu de test :", X_test_pca.shape)

# Affichage du nombre de composantes retenues pour atteindre 95% de variance
print("Nombre de composantes retenues pour 95% de variance :", X_pca_95.shape[1])
```

# 4. Modèles de Classification Supervisée

```python
# Séparation Train / Test
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.20, random_state=42, stratify=y)
```

```python
# Entraînement du modèle KNN
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)

# Prédictions et évaluation
y_pred_knn = knn.predict(X_test)
print("Accuracy KNN :", accuracy_score(y_test, y_pred_knn))
print("Matrice de confusion (KNN) :\n", confusion_matrix(y_test, y_pred_knn))
print("Rapport de classification (KNN) :\n", classification_report(y_test, y_pred_knn))
```

```python
# Entraînement et prédiction avec KNN
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)

# Affichage de la matrice de confusion
cm = confusion_matrix(y_test, y_pred_knn)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=knn.classes_)
disp.plot(cmap='Blues')
plt.title("Matrice de confusion - KNN")
plt.show()
```

```python
# Entraînement du modèle de régression logistique
log_reg = LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=1000, random_state=42)
log_reg.fit(X_train, y_train)

# Prédictions et évaluation
y_pred_logreg = log_reg.predict(X_test)
print("Accuracy Régression Logistique :", accuracy_score(y_test, y_pred_logreg))
print("Matrice de confusion (LogReg) :\n", confusion_matrix(y_test, y_pred_logreg))
print("Rapport de classification (LogReg) :\n", classification_report(y_test, y_pred_logreg))
```

```python
from sklearn.metrics import roc_curve, auc

# Pour un problème à 3 classes, nous allons considérer la classe 1 vs le reste.
# On transforme la variable cible en binaire : 1 si l'étiquette vaut 1, 0 sinon.
y_test_binary = (y_test == 1).astype(int)

# Calcul des probabilités prédites pour la classe 1
y_proba = log_reg.predict_proba(X_test)[:, 1]

# Calcul des taux de vrais positifs (TPR) et faux positifs (FPR)
fpr, tpr, thresholds = roc_curve(y_test_binary, y_proba, pos_label=1)
roc_auc = auc(fpr, tpr)

# Détermination du seuil optimal : on maximise l'indice de Youden (tpr - fpr)
optimal_idx = np.argmax(tpr - fpr)
optimal_threshold = thresholds[optimal_idx]
print("Seuil optimal déterminé (indice de Youden) :", optimal_threshold)

# Tracé de la courbe ROC
plt.figure()
plt.plot(fpr, tpr, color='darkorange',
         label=f'Courbe ROC (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', linestyle='--')  # Ligne de référence
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Taux de Faux Positifs (FPR)')
plt.ylabel('Taux de Vrais Positifs (TPR)')
plt.title('Courbe ROC - Régression Logistique (Classe 1 vs Reste)')
plt.legend(loc="lower right")
plt.show()
```

```python
# Entraînement de l'arbre de décision
dtree = DecisionTreeClassifier(random_state=42)
dtree.fit(X_train, y_train)

# Prédictions et évaluation
y_pred_dt = dtree.predict(X_test)
print("Accuracy Arbre de Décision :", accuracy_score(y_test, y_pred_dt))
print("Matrice de confusion (Arbre de Décision) :\n", confusion_matrix(y_test, y_pred_dt))
print("Rapport de classification (Arbre de Décision) :\n", classification_report(y_test, y_pred_dt))

# Visualisation de l'arbre (simplifié)
plt.figure(figsize=(12,8))
feature_names = data.drop(columns=['target']).columns
plot_tree(dtree, feature_names=feature_names, class_names=['0','1','2'], filled=True, rounded=True)
plt.title("Visualisation de l'Arbre de Décision")
plt.show()
```

# 5. Clustering Non Supervisé avec KMeans

```python
# Application de KMeans avec 3 clusters (en supposant que 3 clusters est optimal)
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X_scaled)

# Calcul de l'indice de silhouette
silhouette = silhouette_score(X_scaled, clusters)
print("Score de silhouette :", silhouette)

# Comparaison des clusters avec la variable cible
cluster_comparison = pd.crosstab(y, clusters, rownames=['Classe réelle'], colnames=['Cluster'])
display(cluster_comparison)

# Méthode du coude pour déterminer le nombre optimal de clusters
inertia = []
k_values = range(1, 11)  # Tester de 1 à 10 clusters
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.figure()
plt.plot(k_values, inertia, marker='o')
plt.xlabel('Nombre de clusters')
plt.ylabel('Inertie')
plt.title('Méthode du coude pour KMeans')
plt.show()
```

```python
# Visualisation des clusters sur la projection PCA 2D
X_pca = X_pca_2
plt.figure()
sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=clusters, palette='viridis')
plt.title("Clusters obtenus par KMeans (projection PCA 2D)")
plt.xlabel("Composante 1")
plt.ylabel("Composante 2")
plt.show()

# Visualisation des clusters sur la projection PCA 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_pca_3[:, 0], X_pca_3[:, 1], X_pca_3[:, 2], c=clusters, cmap='viridis')
ax.set_title("Clusters obtenus par KMeans (projection PCA 3D)")
ax.set_xlabel("Composante 1")
ax.set_ylabel("Composante 2")
ax.set_zlabel("Composante 3")
plt.show()
```

# 6. Optimisation des Hyperparamètres avec Grid Search

```python
# Définition d'une grille de paramètres pour KNN
param_grid_knn = {
    'n_neighbors': [3, 5, 7],
    'weights': ['uniform', 'distance']
}

# GridSearch pour KNN
grid_knn = GridSearchCV(KNeighborsClassifier(), param_grid_knn, cv=5, scoring='accuracy', n_jobs=-1)
grid_knn.fit(X_train, y_train)

# Affichage des meilleurs paramètres et score de validation
print(f"Meilleurs paramètres pour KNN : {grid_knn.best_params_}")
print(f"Score de validation KNN (sur les données d'entraînement) : {grid_knn.best_score_:.4f}")

# Évaluation sur le jeu de test avec les meilleurs paramètres
best_knn = grid_knn.best_estimator_
y_pred_best_knn = best_knn.predict(X_test)
print(f"Accuracy KNN optimisé : {accuracy_score(y_test, y_pred_best_knn):.4f}")
print("Matrice de confusion (KNN optimisé) :\n", confusion_matrix(y_test, y_pred_best_knn))
print("Rapport de classification (KNN optimisé) :\n", classification_report(y_test, y_pred_best_knn))
```

```python
# Définition d'une grille de paramètres pour l'arbre de décision
param_grid_dt = {
    'max_depth': [None, 3, 5],
    'min_samples_split': [2, 5]
}

# GridSearch pour l'arbre de décision
grid_dt = GridSearchCV(DecisionTreeClassifier(random_state=42), param_grid_dt, cv=5, scoring='accuracy', n_jobs=-1)
grid_dt.fit(X_train, y_train)

# Affichage des meilleurs paramètres et score de validation pour l'arbre de décision
print(f"Meilleurs paramètres pour l'Arbre de Décision : {grid_dt.best_params_}")
print(f"Score de validation Arbre de Décision (sur les données d'entraînement) : {grid_dt.best_score_:.4f}")

# Évaluation sur le jeu de test avec les meilleurs paramètres
best_dt = grid_dt.best_estimator_
y_pred_best_dt = best_dt.predict(X_test)
print(f"Accuracy Arbre de Décision optimisé : {accuracy_score(y_test, y_pred_best_dt):.4f}")
print("Matrice de confusion (Arbre de Décision optimisé) :\n", confusion_matrix(y_test, y_pred_best_dt))
print("Rapport de classification (Arbre de Décision optimisé) :\n", classification_report(y_test, y_pred_best_dt))
```

# 7. Validation Croisée

```python
from sklearn.model_selection import cross_val_score

# Liste des modèles à évaluer
models = {
    'KNN': KNeighborsClassifier(n_neighbors=5),
    'Logistic Regression': LogisticRegression(multi_class='multinomial', solver='lbfgs', max_iter=1000, random_state=42),
    'Decision Tree': DecisionTreeClassifier(random_state=42)
}

# Application de la validation croisée pour chaque modèle
for name, model in models.items():
    scores = cross_val_score(model, X_scaled, y, cv=5, scoring='accuracy')  # 5 folds

    # Affichage des scores obtenus pour chaque modèle
    print(f"{name} - Scores CV : {scores}")
    print(f"{name} - Moyenne des scores : {scores.mean():.4f}, Écart-type : {scores.std():.4f}")

    # Discussion sur la stabilité des performances
    if scores.std() > 0.1:
        print(f"{name} montre une grande variabilité dans les scores, ce qui peut indiquer une instabilité du modèle.")
    else:
        print(f"{name} montre des performances stables.\n")

```
